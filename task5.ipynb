```python
#Setting maximum number of columns and rows to display all fields in the output
pd.set_option('display.max_columns', 110)
pd.set_option('display.max_rows', 1100)
```

```python
#reading the data
data=pd.read_csv(r"C:\Users\Maulik\Downloads\heart.csv")
```

```python
#Understanding the data
data.info()
```

```python
#Preview the data
data.head()
```

```python
#Histograms of all the columns
data.hist(figsize=(13,13), bins=40, grid=False, color='skyblue')  
plt.show()
```
